```python
import torch
import cv2
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from moviepy import ImageSequenceClip
import os 

import seaborn as sns
from pathlib import Path
import os

# Set style for better visualizations
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
dataset_path = "dataset_weight_estimation\\dataset_weight_estimation"
vid_carry = cv2.VideoCapture(dataset_path + "\\empty\\ope1\\carry\\angle3.mp4")
vid_walk  = cv2.VideoCapture(dataset_path + "\\empty\\ope1\\walk\\angle1.mp4")
```

```python
total_frames_c = []
total_frames_w = []
for label in ["empty","heavy","light"]:
    for i in range(8) : 
        for j in range(3) : 
            vid_carry = cv2.VideoCapture(dataset_path + f"\\{label}\\ope{i+1}\\carry\\angle{j+1}.mp4")
            frames_carry = []
            while True :
                ret_c,frame_c = vid_carry.read()
                if not ret_c :
                    break   
                frames_carry.append(frame_c)  
            total_frames_c.append(len(frames_carry)) 
```

```python
print(len(total_frames_c))
print(total_frames_c)
```

```python
plt.hist(total_frames_c, bins=10)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution of Values - CARRY VIDEOS")
plt.show()
```

```python
total_frames_w = []
for label in ["empty","heavy","light"]:
    for i in range(8) : 
        for j in range(3) :
            vid_walk  = cv2.VideoCapture(dataset_path + f"\\{label}\\ope{i+1}\\walk\\angle{j+1}.mp4")
            frames_walk = []
            while True :
                ret_w,frame_w = vid_walk.read()
                if not ret_w :
                    break
                frames_walk.append(frame_w) 
            total_frames_w.append(len(frames_walk)) 
        
```

```python
print(len(total_frames_w))
print(total_frames_w)
```

```python
plt.hist(total_frames_w, bins=10)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution of Values - WALK VIDEOS")
plt.show()
```
